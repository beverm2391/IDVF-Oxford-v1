```python
paths = [p for p in os.listdir(DATA_DIR) if p.endswith(".csv")]
assert len(paths) == 93
```

```python
dfs = [pd.read_csv(os.path.join(DATA_DIR, p), index_col=0) for p in paths]
```

## Some test processing

```python
def get_aggs(dfs, interval):
    aggs_df = pd.DataFrame()
    allowed_intervals = {'10T', '30T', '65T'}
    assert interval in allowed_intervals, f"Interval must be one of {[allowed_intervals]}"
    for df in dfs:
        ticker = df['ticker'].iloc[0]
        df['datetime'] = pd.to_datetime(df['t'], utc=True, unit='ms').dt.tz_convert('America/New_York')
        # assert df['datetime'].is_monotonic_increasing, "Datetime column is not monotonically increasing"
        df.set_index('datetime', inplace=True)
        agg = df.resample(interval).agg({'c' : 'mean'})
        agg.columns.values[0] = f"{ticker}"
        aggs_df = pd.concat([aggs_df, agg], axis=1)

    return aggs_df
```

```python
aggs_df = get_aggs(dfs, '65T')
```
