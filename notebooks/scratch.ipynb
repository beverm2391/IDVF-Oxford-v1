```python
def get_ticker(ticker):
    return result[[col for col in result.columns if ticker in col]]
```

```python
tmo = get_ticker('TMO')
```

```python
def plot(df):
    fig, ax = plt.subplots(figsize=(20, 10))

    # Generate a unique color map
    colors = plt.cm.viridis(np.linspace(0, 1, len(df.columns)))

    for i, col in enumerate(df.columns):
        ax.plot(df.index, df[col], label=col, color=colors[i])

    # Additional plot settings
    ax.set_title('Visualization of 93 Columns Over Time')
    ax.set_xlabel('Time')
    ax.set_ylabel('Value')
    ax.legend(loc='upper right', ncol=5)  # Adjust ncol as needed

    plt.show()
```
